```python
import pandas as pd
pok=pd.read_csv("pokemon.csv", squeeze=True).dropna(how="all")
pok["HP"]=pok["HP"].fillna(0).astype(int)
pok["Type 2"]=pok["Type 2"].fillna(0)
pok.head()
```

```python
pok["HP"].rank(ascending=False)
```

```python
pok["HP2"]=pok["HP"].rank(ascending=False).astype("int")
pok.sort_values(by="HP").head()
```
