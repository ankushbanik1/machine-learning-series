```python
import pandas as pd

df=pd.read_csv("100 Records.csv")
df

df.sort_values('First Name',inplace=True)
df.head()
```

```python
mask=df["First Name"].duplicated()
df[mask].head()
```

```python
mask=df["First Name"].duplicated(keep="last")
df[mask].head()
```

```python
mask=~df["First Name"].duplicated(keep=False)
df[mask].head()
```
